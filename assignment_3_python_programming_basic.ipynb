##### 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
number = float(input("Enter a number(float): "))
if number == 0:
    print("Your enter Value is Zero.")
elif number > 0:
    print("Your Enter Value is Positive Number.")
else :
    print("Your Enter Value is Negetive Number.")
```

##### 2. Write a Python Program to Check if a Number is Odd or Even?

```python
number = float(input("Enter a number(float): "))
if (number % 2) == 0:
    print("Your enter Value is Even.")
else:
    print("Your enter Value is Odd.")
```

##### 3. Write a Python Program to Check Leap Year?

```python
year = int(input("Enter a year(yyyy): "))
if (year % 4) == 0:
    print("Your enter year is Leap Year: ",year,"(yyyy)")
else:
    print("Your enter year is not Leap Year: ",year, "(yyyy)")
```

##### 4. Write a Python Program to Check Prime Number?

```python
num = int(input("Enter the Number: "))
if num >= 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, ": not a prime number")
            print(i, "times", num//i, "is", num)
            break
    else:
        print(num, ": prime number")
        
else:
    print(num,": not a prime number")
    
```
